```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.tree import DecisionTreeClassifier,export_graphviz
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
# 1. obtaining data
data = pd.read_csv('developer.csv')
# 2. cleansing data
data.dropna()
data['bug_rate'] = [float(data['bugs'][i]/data['commits'][i]) for i in range(len(data))]
data['fixes_rate'] = [data['fixes'][i]/data['commits'][i] for i in range(len(data))]
data = data[data['commits'] >= 30]

# 3. preprocessing
data_0 = data[data['bugs'] == 0]
data_0['quality'] = 'good'
data_1 = data[data['bug_rate'] > 0]
cnt = 0
br = data_1['bug_rate']

br_list = br.values.tolist()
br_list = sorted(br_list)

# len(data_1)
data_1.loc[br < br_list[int(len(data_1)/3)],'quality'] = 'good'
data_1.loc[br >= br_list[int(len(data_1)*2/3)] ,'quality'] = 'bad'
data_1 = data_1.fillna('middle')
data_1
data_1['quality'].value_counts()

good_data = data_0.append(data_1[data_1['quality'] == 'good'])
good_data = good_data.sample(316,random_state = None)

sample = good_data.append(data_1[data_1['quality'] != 'good'])
# sample = data_1

sample['quality'].value_counts()

```

```python
x = sample[['test','review','report','fixes_rate']]
y = sample['quality']
```

```python
# 3. 数据集划分
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=22)
```

```python
# # 4. 特征预处理
# transfer = StandardScaler()
# x_train = transfer.fit_transform(x_train)
# x_test = transfer.transform(x_test)
```

```python
# 5. 训练模型
estimator = DecisionTreeClassifier(criterion='entropy')
model = estimator.fit(x_train,y_train)
```

```python
# 6. 模型评估
# 1) 直接比对真实值和预测值
y_predict = model.predict(x_test)
# print('y_predict:\n',y_predict)
print('直接比对真实值和预测值：\n',y_test == y_predict)
```

```python
# 2) 计算准确率
score = model.score(x_test,y_test)
print('准确率为：',score)

```

```python

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets

seed=42
kfold=KFold(n_splits=10,random_state=seed)
dtree=DecisionTreeClassifier(criterion='gini',max_depth=10)
dtree=dtree.fit(x_train,y_train)
result=cross_val_score(dtree,x_test,y_test,cv=kfold)
print('决策树结果:',result.mean())
model=AdaBoostClassifier(base_estimator=dtree,n_estimators=100,random_state=seed)
result=cross_val_score(model,x_test,y_test,cv=kfold)
print("提升法改进结果:",result.mean())
```
